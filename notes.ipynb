Create a systemd service unit file:

Create a new service unit file (replace <your-service-name> with a meaningful name for your service, and <container-name> with the name or ID of your Docker container):

bash
Copy code
sudo nano /etc/systemd/system/<your-service-name>.service
Add the following contents to the file:

ini
Copy code
[Unit]
Description=<Description for your service>

[Service]
Restart=always
ExecStart=/usr/bin/docker start -a <container-name>
ExecStop=/usr/bin/docker stop -t 2 <container-name>

[Install]
WantedBy=multi-user.target
Save the file and exit the text editor.

Reload systemd:

After creating the service unit file, you need to reload the systemd manager configuration to make it aware of the new service:

bash
Copy code
sudo systemctl daemon-reload
Enable the service:

To enable the service so that it starts on boot:

bash
Copy code
sudo systemctl enable <your-service-name>.service
Start the service:

You can either manually start the service immediately or reboot your Raspberry Pi to see it start automatically:

bash
Copy code
sudo systemctl start <your-service-name>.service

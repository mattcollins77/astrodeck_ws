in workspace dir

// BUILD IMAGE
docker build -f pi5.Dockerfile -t piimage .

//this runs a container and mounts the volume
docker run -it --name PiContainer --network host --privileged -e DISPLAY=$DISPLAY -v /tmp/.X11-unix:/tmp/.X11-unix -v $(pwd)/src:/home/astrodeck_ws/src -v /dev/:/dev/ piimage

entry.sh will run colcon build and can put in there what pacakges or laucnh files to run

if already got a container
docker start PiContainer
docker exec -it PiContainer /bin/bash

to purge old ones - docker container prune -f
